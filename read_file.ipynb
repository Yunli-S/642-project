```python
def remove_duplicate(name):
    path = "./train/" + name
    df = pd.read_csv(path)
    if 'num_group1' in df.columns:
        df = df.drop(columns=['num_group1'])
    if 'num_group2' in df.columns:
        df = df.drop(columns=['num_group2'])
    without_duplicate_df = df.drop_duplicates(subset=['case_id'], keep="first")
    path_save = "./train_without_duplicate/" + name
    without_duplicate_df.to_csv(path_save, index=False)
    return None
```
